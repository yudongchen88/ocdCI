```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%config Completer.use_jedi = False
```

```python
counts = np.array([[2,1,1,0,7,3,4,3,0,2,2],
                    [0,1,0,1,0,2,4,2,2,1,3],
                    [1,0,0,1,4,1,1,1,0,1,0], 
                    [2,4,2,1,3,3,3,3,3,2,2]])
```

```python
counts
```

```python
industries = ['Communication\nServices',
               'Consumer\nDiscretionary',
               'Consummer\nStaples',
               'Energy',
               'Financials',
               'Health Care',
               'Industrials',
               'Information\nTechnology',
               'Materials',
               'Real Estate',
               'Utilities']
industries
```

```python
dates = ['[2007-02-20, 2007-02-27]', '[2007-05-17, 2007-05-24]', '[2007-07-12, 2007-07-24]', '[2007-08-08, 2007-08-08]']
dates
```

```python
numstocks = np.array([10, 28, 23, 13, 43, 30, 32, 30, 15, 9, 21])
numstocks
```

```python
counts / numstocks
```

```python
ax = sns.heatmap(counts/numstocks, xticklabels=industries, yticklabels=dates, cmap = 'Reds', square=True)
ax.xaxis.set_ticks_position('top')
ax.tick_params(axis='x', rotation=90)
ax.tick_params(length=0)
cbar = ax.collections[0].colorbar
cbar.set_ticks([0, 0.05, 0.1, 0.15, 0.2])
```

```python
plt.figure(figsize = (14,3.5))
ax = sns.heatmap(counts/numstocks, xticklabels=industries, yticklabels=dates, cmap = 'Reds', square=True)
ax.tick_params(length=0)
cbar = ax.collections[0].colorbar
cbar.set_ticks([0, 0.05, 0.1, 0.15, 0.2])
```

```python
figure = ax.get_figure()    
```
